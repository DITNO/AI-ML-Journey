1. End to end scikit-learn workflow

```python
import numpy  as np
```

```python
import pandas  as  pd
heart_disease = pd.read_csv("../Dataset/heart-disease.csv")
heart_disease
```

```python
#create x for features matrix
x = heart_disease.drop("target", axis =1)

#create y (labels)
y = heart_disease["target"]
```

```python
# 2. choose the right model and  hyper parameter
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

#keep the default hyper parameters
clf.get_params()
```

```python
# fit the model to the data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
clf.fit(x_train,y_train);
```

```python
#3. make the prediction
y_label = clf.predict(np,array([0,2,3,4]))
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
#4. evaluate the model on tranning data and test data
clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix , accuracy_score

print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
#5. improve model
# try different amount of parameters

np.random.seed(42)

best_score = 0
best_clf = None

for i in range(10, 100, 10):
    print(f'trying model with the {i} estimators')
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    
    current_score = clf.score(x_test, y_test)
    print(f'model accuracy on test: {current_score*100:.2f}%')
    
    # Track the highest scoring model
    if current_score > best_score:
        best_score = current_score
        best_clf = clf


```

```python
#6. save the model and load it
import pickle

# Save the best tracking model
pickle.dump(best_clf, open("random_forset_model_1.plk", "wb"))
```

```python
loaded_model = pickle.load(open("random_forset_model_1.plk","rb"))
loaded_model.score(x_test,y_test)
```
