```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
car_sales = pd.read_csv("../Dataset/car-sales-extended.csv")
car_sales.head()
# Clean up missing data before splitting or transforming
car_sales = car_sales.dropna()
```

```python
#spilliting the data
x = car_sales.drop("Price", axis=1)
y = car_sales["Price"]
```

```python
categorical_features = ["Make" , "Colour","Doors"]
#also put doors into this as the doors put the category and we all all data  not with category
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("One_hot", one_hot, categorical_features)], remainder="passthrough")

transformer_x = transformer.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(transformer_x,y, test_size = 0.2)

model = RandomForestRegressor(n_estimators=200 , max_depth=10, random_state=42)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error

# 1. Make predictions on the test data
y_preds = model.predict(x_test)

# 2. Calculate the average dollar amount the model is off by
mae = mean_absolute_error(y_test, y_preds)

print(f"Model R^2 Score: {model.score(x_test, y_test):.4f}")
print(f"Average Prediction Error (MAE): ${mae:.2f}")
```
